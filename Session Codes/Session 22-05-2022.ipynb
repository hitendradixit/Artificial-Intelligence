CHARTS

```python
import matplotlib.pyplot as plt
```

BAR CHARTS

```python
sales_data = df.groupby('Customer Segment')['Sales', 'Profit', 'Quantity ordered new'].sum()
```

```python
sales_data
```

```python
plt.bar(sales_data.index, sales_data['Sales'])
plt.show()
```

Give names to x axis, y axis and title

```python
plt.bar(sales_data.index, sales_data['Sales'])
plt.title("Customer Segment wise sales")
plt.xlabel("Customer Segment")
plt.ylabel("Total Sales")
plt.show()
```

Adding a grid line

```python
plt.bar(sales_data.index, sales_data['Sales'])
plt.title("Customer Segment wise sales")
plt.xlabel("Customer Segment")
plt.ylabel("Total Sales")
plt.grid(color = 'red', alpha = 0.5, linestyle = '-.', linewidth = 0.5) #alpha is used for transperancy where 0 is minimum and 1 is highest
plt.show()
```

Limit the vlaue of y axis

```python
plt.bar(sales_data.index, sales_data['Sales'])
plt.title("Customer Segment wise sales")
plt.xlabel("Customer Segment")
plt.ylabel("Total Sales")
plt.ylim(100000, 800000)
plt.show()
```

Horizontal Bar Chart

```python
plt.barh(sales_data.index, sales_data['Sales'])
plt.title("Customer Segment wise sales")
plt.xlabel("Total Sales")
plt.ylabel("Customer Segment")
plt.xlim(100000, 800000)
plt.show()
```

Changing the color of bar chart

```python
color_list = ['pink', 'yellow', 'green', 'orange']
plt.bar(sales_data.index, sales_data['Sales'], color = color_list)
plt.title("Customer Segment wise sales")
plt.xlabel("Customer Segment")
plt.ylabel("Total Sales")
plt.ylim(100000, 800000)
plt.show()
```

Format the axis of bar chart

```python
color_list = ['pink', 'yellow', 'green', 'orange']
plt.bar(sales_data.index, sales_data['Sales'], color = color_list)
plt.title("CUSTOMER SEGMENT WISE SALES", color = 'purple', weight = 'bold')
plt.xlabel("CUSTOMER SEGMENT", color = 'red')
plt.ylabel("TOTAL SALES", color = 'red')
plt.ylim(100000, 800000)
plt.show()
```

Formatting the values on axis

```python
color_list = ['pink', 'yellow', 'green', 'orange']
plt.bar(sales_data.index, sales_data['Sales'], color = color_list)
plt.title("CUSTOMER SEGMENT WISE SALES", color = 'purple', weight = 'bold')
plt.xlabel("CUSTOMER SEGMENT", color = 'red')
plt.ylabel("TOTAL SALES", color = 'red')
plt.xticks(color = 'brown',  rotation = 35)
plt.yticks(color = 'brown')
plt.ylim(100000, 800000)
plt.show()
```

Adding texture to chart

```python
plt.style.use('dark_background')
plt.bar(sales_data.index, sales_data['Sales'], hatch = ('\\'))
plt.title("Customer Segment wise sales")
plt.xlabel("Customer Segment")
plt.ylabel("Total Sales")
plt.ylim(100000, 800000)
plt.show()
```

Adding multiple plots

```python

plt.subplot(2,1,1)
plt.bar(sales_data.index, sales_data['Sales'], hatch = ('\\'))
plt.title("Customer Segment wise sales")
plt.xlabel("Customer Segment")
plt.ylabel("Total Sales")

plt.subplot(2,1,2)
plt.bar(sales_data.index, sales_data['Quantity ordered new'])
#plt.title("Customer Segment wise Quantity ordered new")
plt.xlabel("Customer Segment")
plt.ylabel("Total Quantity ordered new")
plt.legend(sales_data.index, loc = 'center')
plt.show()
```

PIE CHART

```python
sales_data
```

```python
plt.pie(sales_data['Sales'], labels = sales_data.index)
plt.show()
```

Change the color of pie chart

```python
color = ['cyan', 'orange', 'red', 'purple']
plt.pie(sales_data['Sales'], labels = sales_data.index, colors = color )
plt.show()
```

Adding percentage of each pie

```python
color = ['cyan', 'orange', 'red', 'purple']
plt.pie(sales_data['Sales'], labels = sales_data.index, colors = color, autopct = '%1.2f%%' )
plt.show()
```

Slice out

```python
color = ['cyan', 'orange', 'red', 'purple']
explode_list = (0.1,0.1,0.1,0.1)
plt.pie(sales_data['Sales'], labels = sales_data.index, colors = color, autopct = '%1.1f%%', explode = explode_list )
plt.show()
```

Changing the label distance and rotating it

```python
color = ['cyan', 'orange', 'red', 'purple']
explode_list = (0.1,0.1,0.1,0.1)
plt.pie(sales_data['Sales'], labels = sales_data.index, colors = color, autopct = '%1.1f%%', labeldistance = 1.1, rotatelabels = True)
plt.show()
```
